## Import Required Libraries
---

```python
import os
from datetime import datetime
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

## Data Gathering
---

```python
path = r"Dataset\Unemployment.csv"
df = pd.read_csv(path)
df.head()
```

## Preprocessing
---

```python
df.info()
```

The names of the feature starts with `space` which is not ideal. Hence we need to change it to appropriate name using `df.rename({old_column_name : new_column_name} , axis = 1)`

```python
col_names = {'Region' : 'State' , ' Date' : 'Date' , ' Frequency' : 'Frequency' , 
             ' Estimated Unemployment Rate (%)' : 'Unemployment_Rate' , 
             ' Estimated Employed' : 'Employed' , ' Estimated Labour Participation Rate (%)' : 'Labour_Participation_Rate'}
```

```python
df.rename(col_names , axis = 1 , inplace = True)
```

###### Value Counts

```python
df['Frequency'].value_counts()
```

There are some missing values in `Date` Column of df1

```python
df.isna().sum()
```

Removing `Missing Values` using `df.dropna()`

```python
df.dropna(axis = 0 , inplace = True)
```

```python
df.isna().sum()
```

```python
def clean_and_convert(date_str):
    
    date_str = date_str.strip()  # Remove leading and trailing spaces
    
    return pd.to_datetime(date_str, format = '%d-%m-%Y', errors = 'coerce')
```

```python
df['Date'] = df['Date'].apply(clean_and_convert)
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

Now we have completely clean data with `zero` missing values.

```python
df.info()
```

```python
df['Frequency'].value_counts()
```

```python
df['Frequency'].replace({' Monthly' : 'Monthly'} , inplace = True)
```

```python
df['Frequency'].value_counts()
```

## Data Visualization
---

```python
# Histogram of the Unemployment Rate

plt.figure(figsize = (8, 6))

sns.histplot(df['Unemployment_Rate'] , bins = 20 , kde = True)

plt.title('Histogram of Unemployment Rate')

plt.xlabel('Unemployment Rate')

plt.ylabel('Frequency')

plt.show()
```

```python
# Line plot of Unemployment Rate over time (Date)

plt.figure(figsize = (10, 6))

sns.lineplot(x = 'Date', y = 'Unemployment_Rate', data = df)

plt.title('Unemployment Rate over Time')

plt.xlabel('Month')

plt.ylabel('Unemployment Rate')

plt.xticks(rotation = 45)

plt.show()
```

There is rise in unemployment from Month of `March` Till `July` of Year `2020`

```python
monthly_avg_unemployment = df.groupby('Month')['Unemployment_Rate'].mean()

# Plotting the monthly average of Unemployment Rate
plt.figure(figsize=(8, 6))

sns.barplot(x = monthly_avg_unemployment.index, y = monthly_avg_unemployment.values)

plt.title('Monthly Average Unemployment Rate')

plt.xlabel('Month')

plt.ylabel('Average Unemployment Rate')

plt.show()
```

It is seen from the graph that Unemployment is at it's peak in month of `April`, `May` and `June`

```python
numerical_columns = ['Unemployment_Rate', 'Employed', 'Labour_Participation_Rate']

sns.pairplot(df[numerical_columns])

plt.title('Pairplot of Numerical Variables')

plt.show()
```

```python
avg_unemployment = df.groupby('Area')['Unemployment_Rate'].mean()

# Plotting the monthly average of Unemployment Rate
plt.figure(figsize=(8, 6))

sns.barplot(x = avg_unemployment.index, y = avg_unemployment.values)

plt.title('Area wise Average Unemployment Rate')

plt.xlabel('Area')

plt.ylabel('Average Unemployment Rate')

plt.show()
```

```python
state_unemployment = df.groupby('State')['Unemployment_Rate'].mean().reset_index()

# Sort the data in descending order to find the state with the highest unemployment rate
state_unemployment_sorted = state_unemployment.sort_values(by = 'Unemployment_Rate', ascending=False)

# Plot the bar plot
plt.figure(figsize = (12, 6))

sns.barplot(x = 'State', y = 'Unemployment_Rate', data = state_unemployment_sorted)

plt.xticks(rotation = 90)

plt.xlabel('State')

plt.ylabel('Average Unemployment Rate')

plt.title('Average Unemployment Rate by State')

plt.tight_layout()

plt.show()
```

```python
state_area_df = df.groupby(['Area', 'State']).agg({'Unemployment_Rate': 'mean'}).reset_index()
sns.scatterplot(state_area_df , x = 'State' , y = 'Unemployment_Rate' , hue = 'Area')

plt.xticks(rotation = 90)

plt.xlabel('State')

plt.ylabel('Average Unemployment Rate')

plt.title('Average Unemployment Rate by State Area')

plt.tight_layout()

plt.show()
```

```python
avg_employment = df.groupby('Area')['Employed'].mean()

# Plotting the average of employed against Area
plt.figure(figsize=(8, 6))

sns.barplot(x = avg_employment.index, y = avg_employment.values)

plt.title('Area wise Average Employed')

plt.xlabel('Area')

plt.ylabel('Average Employed')

plt.show()
```

```python
state_employment = df.groupby('State')['Employed'].mean().reset_index()

# Sort the data in descending order to find the state with the highest employed
state_employment_sorted = state_employment.sort_values(by = 'Employed', ascending=False)

# Plot the bar plot
plt.figure(figsize = (12, 6))

sns.barplot(x = 'State', y = 'Employed', data = state_employment_sorted)

plt.xticks(rotation = 90)

plt.xlabel('State')

plt.ylabel('Average Employed')

plt.title('State Wise Average Employed')

plt.tight_layout()

plt.show()
```

```python
state_area_df = df.groupby(['Area', 'State']).agg({'Employed': 'mean'}).reset_index()
sns.scatterplot(state_area_df , x = 'State' , y = 'Employed' , hue = 'Area')

plt.xticks(rotation = 90)

plt.xlabel('State')

plt.ylabel('Average Employed')

plt.title('State Area Wise Average Employed')

plt.tight_layout()

plt.show()
```

### Checking Correlation

```python
def correlation_plot(data: pd.DataFrame = None, numeric_only: bool = True, width: int = 10, height: int = 5):
    
    corr = data.corr(numeric_only=numeric_only)
    
    mask_ = np.array(corr)
    mask_[np.tril_indices_from(mask_)] = False

    fig , ax = plt.subplots()
    fig.set_size_inches(width,height)
    sns.heatmap(corr , mask = mask_ , vmax = 0.8 , square = True , annot = True , cmap = "YlGnBu")
    plt.show()
```

```python
correlation_plot(df)
```
